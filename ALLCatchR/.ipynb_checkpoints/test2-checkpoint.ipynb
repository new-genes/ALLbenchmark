```r
# 필요한 패키지 로드
library(readr)
```

```r
# example.txt 파일을 읽어들입니다.
data <- read.table("/bce/groups/pnucolab/analysis/Bohun_Ranking_Pathway/Bohun_Shares/Expression_Profile_240507.txt", header = TRUE, sep = "\t", stringsAsFactors = FALSE)

# 필요한 열만 선택하여 새로운 데이터 프레임을 만듭니다.
selected_data <- data[, c("Gene_Symbol", "R1_Read_Count", "R2_Read_Count", "R4_Read_Count")]

# 데이터 출력
head(selected_data)
```

```r
# 중복된 값 확인 및 해당 행 제거
df_unique <- selected_data[!duplicated(selected_data$Gene_Symbol) & !duplicated(selected_data$Gene_Symbol, fromLast = TRUE), ]

# 중복 제거된 데이터 프레임 출력
print("Data Frame with Duplicates Removed")
print(df_unique)
```

```r
rownames(df_unique) <- df_unique$Gene_Symbol
df_unique
```

```r
df_unique <- df_unique[,2:ncol(df_unique)]
df_unique
```

```r
# 텍스트 파일로 저장
write.table(df_unique, file = "./input2.txt", sep = "\t")  
# 출력 경로와 출력 파일 이름을 원하는 위치와 파일 이름으로 수정해주세요
# sep = "\t"는 탭으로 구분된 파일을 생성하기 위한 설정입니다. 필요에 따라 다른 구분자로 수정할 수 있습니다.
```

```r
out <- allcatch_bcrabl1(Counts.file = "input2.txt",  ID_class = "symbol", sep = "\t", out.file = "predictions2.tsv")
```
