```r
# CSV 파일 읽기
data <- read.csv("/bce/groups/pnucolab/analysis/Bohun_Ranking_Pathway/Bohun_Shares/total_fpkm_uq.csv", header = TRUE)
data
```

```r
# 두 번째 열을 제외한 나머지 열 선택
data_selected <- data[,2:ncol(data)]  # 첫 번째 열을 제외한 나머지 열 선택
data_selected
```

```r
# 중복된 값 확인 및 해당 행 제거
df_unique <- data_selected[!duplicated(data_selected$gene_name) & !duplicated(data_selected$gene_name, fromLast = TRUE), ]

# 중복 제거된 데이터 프레임 출력
print("Data Frame with Duplicates Removed")
print(df_unique)
```

```r
# 지정한 열을 정수형으로 변환
df_unique$SJALL014946_D1 <- as.integer(df_unique$SJALL014946_D1)
df_unique$SJALL014947_D1 <- as.integer(df_unique$SJALL014947_D1)
df_unique$SJALL014949_D1 <- as.integer(df_unique$SJALL014949_D1)
df_unique$SJALL014950_D1 <- as.integer(df_unique$SJALL014950_D1)
df_unique$SJALL014952_D1 <- as.integer(df_unique$SJALL014952_D1)
```

```r
rownames(df_unique) <- df_unique$gene_name
df_unique
```

```r
df_unique <- df_unique[,2:ncol(df_unique)]
df_unique
```

```r
# 텍스트 파일로 저장
write.table(df_unique, file = "./input.txt", sep = "\t")  
# 출력 경로와 출력 파일 이름을 원하는 위치와 파일 이름으로 수정해주세요
# sep = "\t"는 탭으로 구분된 파일을 생성하기 위한 설정입니다. 필요에 따라 다른 구분자로 수정할 수 있습니다.
```

```r
out <- allcatch_bcrabl1(Counts.file = "input.txt",  ID_class = "symbol", sep = "\t", out.file = "predictions.tsv")
```
